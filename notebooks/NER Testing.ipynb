```python
#Initialize matcher
matcher = Matcher(nlp.vocab)

#Define pattern
pattern = [{"IS_DIGIT": True},
    {"TEXT": "block"}]

#Add the pattern to the matcher
matcher.add("block", [pattern])

doc = nlp('''
MEXICO, Mo. (KMIZ) The Mexico Missouri Department of Public Safety reported two fires that occurred early Friday morning.  No injuries were reported  The first was a fire at an apartment complex in the 1200 block of West Breckenridge Drive.  There was extensive damage to 11 apartments and the complex office, according to the department of safety. The fire is still under investigation. Another fire was reported at 3:41 a.m. in the 1000 block of Elmwood Drive. Fire was found in the building. That fire is also under investigation.
''')
matches = matcher(doc)

for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(span.text)

```

```python
#Common Street endings and abbreviations

#Getting the list from the USPS
#https://www.expertmarket.com/postage-meters/usps-street-suffix-abbreviations

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

url = "https://www.expertmarket.com/postage-meters/usps-street-suffix-abbreviations"

response = requests.get(url)

USPS_soup = BeautifulSoup(response.text, "html")

all_tables=USPS_soup.find_all('table', class_='tableizer-table')

usps_tables = all_tables

common_streets = []

for row in USPS_soup.findAll("td")[1:]: 
    if row.text != '':
        common_streets.append(row.text.lower())
        
#Remove duplicates
common_streets_final = set(common_streets)
common_streets_final = list(common_streets_final)
```
